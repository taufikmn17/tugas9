```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

dataku = pd.read_csv('halamanharga.csv')

from sklearn.preprocessing import StandardScaler

# Menormalisasi
scr = StandardScaler()
dataku_norm = scr.fit_transform(dataku[['pages','price']])
dataku_norm = pd.DataFrame(dataku_norm, columns=['pages','price'])

wcss = []

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(dataku_norm)
    wcss.append(kmeans.inertia_)
    
# Nilai WCSS ke grafik
plt.plot(range(2, 11), wcss)
plt.title('Menggunakan Metode Elbow')
plt.xlabel('Klaster')
plt.ylabel('WCSS')
plt.show()

# Klastering
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(dataku_norm)

dataku_norm['label'] = kmeans.labels_

# grafik hasil klasteringnya
plt.scatter(dataku_norm['pages'][dataku_norm.label == 0], 
            dataku_norm['price'][dataku_norm.label == 0], s = 50, c = 'brown', label = 'Klaster 1')
plt.scatter(dataku_norm['pages'][dataku_norm.label == 1], 
            dataku_norm['price'][dataku_norm.label == 1], s = 50, c = 'green', label = 'Klaster 2')
plt.scatter(dataku_norm['pages'][dataku_norm.label == 2], 
            dataku_norm['price'][dataku_norm.label == 2], s = 50, c = 'yellow', label = 'Klaster 3')
plt.scatter(dataku_norm['pages'][dataku_norm.label == 3], 
            dataku_norm['price'][dataku_norm.label == 3], s = 50, c = 'purple', label = 'Klaster 4')
plt.scatter(dataku_norm['pages'][dataku_norm.label == 4], 
            dataku_norm['price'][dataku_norm.label == 4], s = 50, c = 'blue', label = 'Klaster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'red', label = 'Centroids')
plt.title('Klastering menggunakan K-Means (k=5)')
plt.xlabel('Pages')
plt.ylabel('Price')
plt.legend()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

dataku = pd.read_csv('halamanharga.csv')

from sklearn.preprocessing import StandardScaler

# Menormalisasi
scr = StandardScaler()
dataku_norm = scr.fit_transform(dataku[['pages','price']])
dataku_norm = pd.DataFrame(dataku_norm, columns=['pages','price'])



from sklearn.metrics import silhouette_score
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(dataku_norm)
    silhouette_scores.append(silhouette_score(dataku_norm, kmeans.labels_))

# visualisasi
plt.plot(range(2,11), silhouette_scores)
plt.xlabel('Klaster')
plt.ylabel('Skor Silhoutte')
plt.title('Silhouette')
plt.show()

# klastering
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=42)
kmeans.fit(dataku_norm)

# menambahkan kolom
dataku_norm['label'] = kmeans.labels_

# grafik hasil klasteringnya
plt.scatter(dataku_norm['pages'][dataku_norm.label == 0], 
            dataku_norm['price'][dataku_norm.label == 0], s = 50, c = 'brown', label = 'Klaster 1')
plt.scatter(dataku_norm['pages'][dataku_norm.label == 1], 
            dataku_norm['price'][dataku_norm.label == 1], s = 50, c = 'green', label = 'Klaster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'blue', label = 'Centroids')
plt.title('Klastering (k=5)')
plt.xlabel('Pages')
plt.ylabel('Price')
plt.legend()
```
